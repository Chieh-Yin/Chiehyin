### ♠ Dijkstra 原理 ♠

Dijkstra又稱作戴克斯特拉算法，

使用了廣度優先搜尋，

是尋找從起點到其他節點的最短路徑演算法，

用來解決所謂的最短路徑問題(Shortest Path Problem)。

我認為Dijkstra可以細分成三個部分:

    ➀ 找一個節點當作起點，之後找到與起點相連但尚未被選到過的節點中距離最短的那個節點當下一個
    
    ➁ 並且再從這個選到的節點找到跟他相連的其他節點，並且依樣在取與它距離最短的當作下一個
    
    ➂ 當所有點都被當所有點都被家道過後，Dijkstra就完成了
    


### ♠ Kruskal 原理 ♠

Kruskal又稱作克魯斯克爾算法，

是尋找最小生成樹(Minimum Spanning Tree)的演算法，

而我認為Kruskal一樣可以細分成三個部分:

    ➀ 將各個點連成的邊按照大小排列，從權值最低的邊開始

    ➁ 在這個過程中不能發生迴圈的問題
    
    ➂ 直到加了所有節點-1後停止

### ♠ Dijkstra 與 Kruskal 的時間複雜度 ♠

* Dijkstra
    
    時間複雜度：O(|E|+|V|log|V|)
    
* Kruskal

    時間複雜度：O(|E|log|V|)
    
(V : 節點，E : 邊)
