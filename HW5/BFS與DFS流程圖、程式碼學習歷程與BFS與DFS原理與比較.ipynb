### ♠ BFS 原理 ♠

BFS(Breadth-First Search)稱作"廣度優先搜尋法"，

他是一種圖形搜索演算法，是屬於盲目搜索。

通常利用Queue來處理，並以迴圈的方式進行。

   * Queue是所謂的佇列，
   
     好比排隊出場，一個一個按照先後順序，先進先出(FIFO)，
     
     而這次主要會需要用到的語法包括:
     
     Push(Data) : 將資料加進去Queue後面，並更新成新的Back
     
     Pop : 把Front所指向的資料從Queue中移除，並更新Front，(也可寫dequeue)
     
     
而我認為BFS的步驟可以分成幾個部分講解:

    1. 從出發點開始，走訪所有跟出發點有連結到的節點

    2. 之後再依序走訪與先前走訪過的節點有相連結但尚未被拜訪過的節點

    3. 直到所有節點都已被拜訪過後才算完成BFS


### ♠ DFS 原理 ♠

DFS(Depth-First Search)稱作"深度優先搜尋法"，

他是一種用來搜尋一個樹或圖的演算法，跟BFS一樣亦屬於盲目搜索。

通常利用Stack來處理，並以遞迴的方式進行。

   * Stack是所謂的堆疊，
   
     好比堆盤子，一層一層往上疊，但要拿時卻是最後一個先取出，後進先出(LIFO)，
     
     而這次主要會需要用到的語法包括:
     
     Push(Data) : 將資料加進去Stack

     Pop : 把最上面的數取出來
     
而我認為DFS的步驟可以分成幾個部分講解:

    1. 從出發點開始，走訪所有跟出發點有連結到的節點

    2. 之後再走訪與之前走訪的節點有相連結的節點，並依序放入Stack中

    3. 將最上面的拿出來當作下一個要走訪的節點
    
    4. 重複2、3步驟直到所有節點都被拜訪過後才算完成DFS

### ♠ BFS 與 DFS 之比較 ♠

* BFS

    1. 以迴圈方式進行
    
    2. 用Queue處理
    
    3. 從一個節點開始，將節點一層一層的拜訪下去，深度最淺的節點會先拜訪
    
    4. 需要與節點數成正比的記憶體空間
* DFS

    1. 以遞迴方式進行
    
    2. 用Stack處理
    
    3. 從一個節點開始，一直往尚未拜訪過的節點走去，先遇到的就先拜訪
    
    4. 較省記憶體空間

### ♠ BFS 與 DFS 的複雜度 ♠

* BFS

    時間複雜度：O(|V|+|E|) = O(b^d)

    空間複雜度：O(|V|+|E|) = O(b^d)
    
* DFS

    時間複雜度：O(b^m)

    空間複雜度：O(bm)
    
    (b: 分支係數、m: 圖的最大深度)

### Step 2:

第二個步驟我的想法是，

我要利用stack的方式，將放在state1的點取出，

而因為stack是後進先出，

所以我利用state1[-1]這個方式來取暫時的list中的最後一個點，

並把他加到state2中，然後再刪除。

刪除的方式我一樣也是利用state1[:-1]，

state1變成從第一個點到最後一個點的前面一個為止。

這邊一樣用寫在紙上的那組測值，

做第一次的執行過程。

```python
# Python3 Program to print BFS traversal 
# from a given source vertex. BFS(int s) 
# traverses vertices reachable from s. 
from collections import defaultdict 
  
# This class represents a directed graph 
# using adjacency list representation 
class Graph:
    # Constructor 
    def __init__(self): 
        # default dictionary to store graph 
        self.graph = defaultdict(list) 

    # function to add an edge to graph 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 

        
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        
        state1 = [] #暫時的
        state2 = [] #排好的
        
        state2.append(s)
        state1.extend(self.graph[s])
     
      
        top = state1[-1] #取最後加入的當下一個點
        state1 = state1[:-1] #刪掉最後加入的
        state2.append(top) #加到排好的list當下一個點

        print("state1: ",state1)
        print("state2: ",state2)
```

### ♠ 附上兩部分合併的完整程式碼 ♠

```python
# Python3 Program to print BFS traversal 
# from a given source vertex. BFS(int s) 
# traverses vertices reachable from s. 
from collections import defaultdict 
  
# This class represents a directed graph 
# using adjacency list representation 
class Graph:
    # Constructor 
    def __init__(self): 
        # default dictionary to store graph 
        self.graph = defaultdict(list) 

    # function to add an edge to graph 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
  
    # Function to print a BFS of graph 
    def BFS(self, s): 
        """
        :type s: int
        :rtype: list
        """
        state1 = [] 
        state2 = [] 
        
        state2.append(s)
        state1.extend(self.graph[s])  
        
        while state1  : 
            if state1:
                first = state1[0]
                state1 = state1[1:]
                state2.append(first)
            else:
                pass
            
            for i in self.graph[first]:
                if i not in state1 and i not in state2:
                    state1.append(i)
        return state2
        
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """       
        state1 = []
        state2 = []
        
        state2.append(s)
        state1.extend(self.graph[s])
     
        while state1:
            top = state1[-1]
            state1 = state1[:-1]
            state2.append(top)
            
            for i in self.graph[top]:
                if i not in state1 and i not in state2:
                    state1.append(i)
        return state2
```

```python
g = Graph()
g.addEdge(0,1)
g.addEdge(0,2)
g.addEdge(1,2)
g.addEdge(2,0)
g.addEdge(2,3)
g.addEdge(3,3)

print(g.BFS(2))
print(g.DFS(2))
```
