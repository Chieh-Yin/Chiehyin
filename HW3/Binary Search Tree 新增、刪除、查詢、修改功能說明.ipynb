## ♧ Binart Search Tree 功能說明 ♧

Binary Search Tree主要分成四種功能，

分別是: 新增、刪除、修改、查詢，

以下將分開說明各個功能的作用。

### ♧ 新增 Insert ♧ 

將一個節點插入樹中，其可能產生的情況有:

1. 這棵樹還甚麼東西都沒有，所以直接將要插入的節點當作樹根

2. 如果要插入的節點<=樹根，則往左

3. 如果要插入的節點>樹根，則往右

### ♧ 查詢 Search ♧ 

我們想要查詢一個數，其可能產生的情況有:

1. 這棵樹不存在，回傳None

2. 這棵樹存在，
    
    * 所要查詢的數 = 樹根，直接回傳查詢的值

    * 所要查詢的數 < 樹根，往左找
    
    * 所要查詢的數 > 樹根，往右找
    
     有找到，查詢成功，回傳值；沒找到，查詢失敗。

### ♦ 遍歷:
分成三種方式進行，主要不同之處是順序。
    1. 前序 pre-order: 樹根、左節點、右節點
    2. 中序 in-order: 左節點、樹根、右節點
    3. 後序 post-order: 左節點、右節點、樹根

### ♧ 刪除 Delete ♧ 

想要刪掉一個節點時，可能會產生的情況:

   1. 想要移除的節點下面沒有其他節點:
        * 直接移除
   2. 想要移除的節點左邊有節點:    
        * 移除後將左邊的第一個節點直接取代原本節點的位置
   3. 想要移除的節點右邊有節點:
        * 移除後將右邊的第一個節點直接取代原本節點的位置
   4. 想要移除的節點左右兩邊都還有其他節點:
        * 這邊會存在兩個解法:           
            * 移除後讓右節點中最小的節點(右節點中左節點的最後一項)移上來，取代移除掉的節點的位置
            * 移除後讓左節點中最大的節點(左節點中右節點的最後一項)移上來，取代移除掉的節點的位置
